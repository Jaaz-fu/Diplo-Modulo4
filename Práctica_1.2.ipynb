### Data Transform

Drop Rows with missing `count` 

```python
df_dropped = df.dropna(subset=['revenue'])
df_dropped.head()
```

```python
df_dropped.shape
```

Copy of df

```python
df_copy = df_dropped.copy()
df_copy.head()
```

```python
df_copy.shape
```

### Handling Categorical Data

```python
list(df['status'].unique())
```

###### Create a dictionary

Crearemos el diccionario sobre la columna de 'Genre'

```python
keys = list(df['status'].unique())
vals = range(1,7)
act = dict(zip(keys, vals))
act
```

Creamos una nueva variable donde tengamos los generos como etiquetas numéricas.

```python
df_copy['status_2'] = df['status'].map(act)
df_copy.head()
```

### Data Summarization
Preparing a compact representation of raw data at hand


```python
df_copy.groupby(["original_language","adult", "status",]).count()
```

### Data Visualization

#### Scatter plots

Variables able to be highly correlated. To confirm or validate our hypothesis.

###### Queremos ver como se comportan las ganancias por la duraciòn de las pelìculas

```python
df_copy['runtime'] = df_copy['runtime'].astype('float')

df_copy.plot(kind='scatter',x='runtime',y='revenue')
```

#### Histograms

```python
plt.hist(df_copy['revenue'])
plt.xlabel('frequency')
plt.ylabel('revenue')
plt.title('Frequency of revenue')
```

#### Boxplots

```python
#plt.boxplot(df_sample['ventas'])
plt.boxplot(df_copy['revenue'], showmeans=True, meanline=True)
plt.ylabel('revenue count')
plt.title('Box Plot of revenue count')
```
